```python
#Calculate variance
N = 1000
more = cp6[cp6['numwebvisitsmonth'] > 5]['numwebpurchases'].sample(1000, random_state=42)
less = cp6[cp6['numwebvisitsmonth'] <=5]['numwebpurchases'].sample(1000, random_state=42)
```

```python
more_var = more.var()
print(more_var)
less_var = less.var()
print(less_var)
```

```python
#calculate Standard Deviation
std = np.sqrt((more_var + less_var)/2)
print('std dev: ', std)
```

```python
#calculate test statistic
t = (more.mean() - less.mean()) / (std*np.sqrt(2/N))
print('t: ', t)
```

```python
# Null Hypothesis Dictionary
null_hypothesis = {
    'H0': "Web visits do not contribute to online sales"
    , 'H1': "Web visits contribute to online sales"
}
```

```python
def print_sig(p_value, alpha):
    '''
    Inputs:
    p_value: Calculated p_value
    alpha: Confidence level
    '''
    if p_value < alpha:
        print("We reject our null hypothesis.")
        print(null_hypothesis['H1'])
    elif p_value > alpha:
        print("We fail to reject our null hypothesis.")
        print(null_hypothesis['H0'])
    else:
        print("Our test is inconclusive.")
```

```python
print("t = " + str(t))
print("p = " + str(p*2))
print_sig(p*2, alpha)
```
