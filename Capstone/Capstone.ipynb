```python
ar=auto_arima(ysc, 
                     exogenous= Xsc,
                     start_p=1,  start_q=1, 
                     max_p=7,  max_q=7, 
                     d=1, max_d=7,
                     trace=True, 
                     error_action='ignore', 
                     suppress_warnings=True, 
                     stepwise=True)
```

```python
model= SARIMAX(ysc, 
           exog=Xsc,
              order=(2,1,2),
              enforce_invertibility=False, enforce_stationarity=False)
```

```python
results = model.fit()
```
